```python
import pandas as pd

from joblib import load

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)
```

```python
df = pd.read_csv(
    "../dataset/processed/final_dataset.csv"
)

print(df.shape)
```

```python
encoders = {}

for col in [
    "state",
    "district",
    "crop",
    "season"
]:
    le = LabelEncoder()

    df[col] = le.fit_transform(
        df[col].astype(str)
    )

    encoders[col] = le
```

```python
features = [
    "state",
    "district",
    "crop",
    "season",
    "area",
    "production",
    "population",
    "monthly_rainfall"
]

X = df[features]
```

```python
y_demand = df["demand"]

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X,
    y_demand,
    test_size=0.2,
    random_state=42
)
```

```python
demand_model = load(
    "../models/demand_prediction_model.pkl"
)
```

```python
demand_pred = demand_model.predict(
    X_test_d
)
```

```python
print("MAE:")
print(
    mean_absolute_error(
        y_test_d,
        demand_pred
    )
)

print("\nRMSE:")
print(
    mean_squared_error(
        y_test_d,
        demand_pred
    ) ** 0.5
)

print("\nR2 Score:")
print(
    r2_score(
        y_test_d,
        demand_pred
    )
)
```

```python
y_risk = df["risk_level"]

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X,
    y_risk,
    test_size=0.2,
    random_state=42
)
```

```python
risk_model = load(
    "../models/risk_classification_model.pkl"
)
```

```python
risk_pred = risk_model.predict(
    X_test_r
)
```

```python
print(
    "Accuracy:",
    accuracy_score(
        y_test_r,
        risk_pred
    )
)
```

```python
print(
    classification_report(
        y_test_r,
        risk_pred
    )
)
```

```python
print(
    confusion_matrix(
        y_test_r,
        risk_pred
    )
)
```
